```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
data = {
    "model": ["Decision tree", "Fair Decision tree", "Equalized odds Decision tree", "SVM", "Fair SVM", "Equalized odds SVM", "Neural network", "Fair Neural network"],
    "accuracy": [0.8787, 0.8971, 0.8607, 0.9056, 0.8995, 0.8984, 0.9070, 0.9010],
    "weighted average F1-score": [0.8587, 0.8832, 0.8614, 0.8901, 0.8820, 0.8773, 0.8956, 0.8903]
}

df = pd.DataFrame(data)

df
```

```python
decision_tree_regex = re.compile(r"decision tree", re.IGNORECASE)
svm_regex = re.compile(r"svm", re.IGNORECASE)
neural_networks_regex = re.compile(r"neural network", re.IGNORECASE)

decision_tree_data = df[df["model"].str.contains(decision_tree_regex)]
svm_data = df[df["model"].str.contains(svm_regex)]
neural_networks_data = df[df["model"].str.contains(neural_networks_regex)]

plt.plot(decision_tree_data.index, decision_tree_data["accuracy"], marker='o', label="Decision tree")
plt.plot(svm_data.index, svm_data["accuracy"], marker='o', label="SVM")
plt.plot(neural_networks_data.index, neural_networks_data["accuracy"], marker='o', label="Neural networks")

# Adding labels and title
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.title("Accuracy of Models")

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45)

# Adding a legend
plt.legend()

# Displaying the chart
plt.show()
```

```python
# Define the model categories using regular expressions
decision_tree_regex = re.compile(r"decision tree", re.IGNORECASE)
svm_regex = re.compile(r"svm", re.IGNORECASE)
neural_networks_regex = re.compile(r"neural network", re.IGNORECASE)

# Separate data by model categories
decision_tree_data = df[df["model"].str.contains(decision_tree_regex)]
svm_data = df[df["model"].str.contains(svm_regex)]
neural_networks_data = df[df["model"].str.contains(neural_networks_regex)]

# Plotting the line chart for each model category
plt.plot(decision_tree_data.index, decision_tree_data.iloc[:, 2], marker='o', label="Decision tree")
plt.plot(svm_data.index, svm_data.iloc[:, 2], marker='o', label="SVM")
plt.plot(neural_networks_data.index, neural_networks_data.iloc[:, 2], marker='o', label="Neural networks")

# Adding labels and title
plt.xlabel("Model")
plt.ylabel("Weighted Average F1-score")
plt.title("Weighted Average F1-score of Models")

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45)

# Adding a legend
plt.legend()

# Displaying the chart
plt.show()
```
